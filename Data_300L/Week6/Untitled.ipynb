```python
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
palette = sns.color_palette("bright", n_colors=4)
```

```python
cmap = ListedColormap(sns.color_palette(palette).as_hex())
```

```python
num_dots = 200
```

```python
random_1 = np.random.randn(num_dots)
random_2 = np.random.randn(num_dots)
```

```python
colors = np.random.randint(0, 10, num_dots)
```

```python
plt.scatter(random_1, random_2, c=colors, cmap=cmap)
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))
iris = sns.load_dataset('iris')
sns.swarmplot(x='species', y ='petal_length', data=iris, ax=ax, size=4)
plt.show()
```

```python
tips_data = sns.load_dataset("tips")
ax = sns.boxplot(x="total_bill", data=tips_data)
ax.set_title("boxplot")
plt.show()
```

```python
lm = sns.lmplot(x ='tip', y='total_bill', data =tips_data, height=7, truncate=True)
fig = lm.fig
fig.suptitle('Put the figure title here', fontsize=12)
plt.show()
```

```python
x_vals = 10 ** np.arange(1, 5)
y_vals = x_vals * 5 
data = pd.DataFrame(data={'x': x_vals, 'y': y_vals})
grid = sns.lmplot(x='x', y='y', data=data, height=7, truncate =True)
grid.set_xticklabels(rotation=30)
plt.show()
```

```python
sns.set(style='darkgrid')
a = sns.load_dataset('iris')
b = sns.FacetGrid(a, col='species')
b.map(plt.hist, 'sepal_length')
plt.show()
```
