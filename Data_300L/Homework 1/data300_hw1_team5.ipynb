```python
import matplotlib.pyplot as plt
import pandas as pd

sunshine = pd.read_csv('/Users/connorjohnson/Desktop/Data_300L/Homework 1/sunshine.csv')

fig1 = plt.figure(figsize=(12,5))

ax = plt.gca()

sunshineInCity = sunshine.pivot(index='month', columns='city', values='sunshine')

# Reorder columns of the dataframe
sunshineInCity = sunshineInCity[['Seattle','Chicago','New York', 'San Francisco', 'Houston', 'Miami']]

# plot the results using a line plot
colors = ['#98d4f2', '#14acff', '#006faf',  '#ffbcbc', '#ff3d3d', '#ce0000']

for index, column in enumerate(sunshineInCity.columns):
    ax.plot(sunshineInCity[column], label=sunshineInCity[column].name, color=colors[index])

# xticks size
ax.tick_params(axis='both', labelsize = 14)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# legend 
ax.legend(fontsize=14, bbox_to_anchor=(1.05, 1), loc='best')

# Text
plt.text(-2.2, 360, 'What is the Difference Between', fontsize=20, color='grey')
plt.text(3.67, 360, 'Northern', fontsize=20, color='#74cbf7')
plt.text(5.38, 360, 'and', fontsize=20, color='grey')
plt.text(6.18, 360, 'Southern', fontsize=20, color='red')
plt.text(7.95,360,'City Monthly Sunshine Levels?', fontsize = 20, color = 'grey')

ax.set_xlabel('Months', fontsize=16)
ax.set_ylabel('Sunshine', fontsize=16)
```
