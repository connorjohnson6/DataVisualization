```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
covid = pd.read_csv('/Users/connorjohnson/Desktop/Data_300L/Homework 2/Covid.csv')
```

```python
high_risk_percentage = covid["total_percent_at_risk"]
MMSA = covid["MMSA"]

# Sort the data in ascending order based on high_risk_percentage
sorted_indices = high_risk_percentage.argsort()
sorted_high_risk_percentage = high_risk_percentage[sorted_indices]
sorted_MMSA = MMSA[sorted_indices]

# Create a bar chart
plt.bar(sorted_MMSA, sorted_high_risk_percentage, color="#3D5A80")

# Set the plot title and axis labels
plt.title("Total Percentage of High-Risk Individuals in Each Location")
plt.xlabel("Location")
plt.ylabel("Total Percentage of High-Risk Individuals")

plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
# create a pandas dataframe with the hospital information
hospitals = pd.DataFrame({
    'Hospital': ['Hilton Head Island-Bluffton-Beaufort, SC', 'Cambridge-Newton-Framingham, MA, Metropolitan Division',
                'Riverside-San Bernardino-Ontario, CA', 'New York-Jersey City-White Plains, NY-NJ, Metropolitan Division',
                'Sioux Falls, SD', 'Austin-Round Rock, TX', 'Miami-Fort Lauderdale-West Palm Beach, FL',
                'Chicago-Naperville-Elgin, IL-IN-WI', 'North Platte, NE', 'Reno, NV'],
    'Region': ['South', 'North', 'South', 'North', 'North', 'South', 'South', 'North', 'North', 'South'],
    'Total Percent at Risk': [62.72, 52.17, 56.62, 54.38, 58.02, 51.27, 55.77, 56.02, 64.19, 57.86]
})

# create a line graph of the total percentage at risk by hospital, with colors based on region
for region, color in [('South', 'red'), ('North', 'blue')]:
    df = hospitals[hospitals['Region'] == region]
    plt.plot(df['Hospital'], df['Total Percent at Risk'], color=color, label=region)

# set the axis labels and title
plt.xlabel('Hospital')
plt.ylabel('Total Percent at Risk')
plt.title('Regional Differences in Percentage of High-Risk Individuals for COVID-19')

plt.xticks(rotation=90)
# add a legend to the chart
plt.legend()

# display the chart
plt.show()

```

```python
sns.set_style("darkgrid")

# create the scatter plot with hue
sns.relplot(x="total_at_risk", y="icu_beds", data=covid, kind='scatter', hue='MMSA')

# add a title
plt.title("ICU Beds vs Total At Risk")

# show the plot
plt.show()
```

```python
# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(15, 5))

# Create the boxplot with the desired color and x-axis label
sns.boxplot(x="high_risk_per_hospital", data=covid, color="#ef5858", ax=ax)
ax.set_title("High Risk Per Hospital Boxplot")
ax.set_xlabel("Number of High Risk Per Hospital")

# Show the plot
plt.show()
```

```python
# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(15, 5))

# Create the boxplot with the desired color and x-axis label
sns.boxplot(x="high_risk_per_ICU_bed", data=covid, color="#ef5858", ax=ax)
ax.set_title("High Risk Per ICU Boxplot")
ax.set_xlabel("Number of High Risk Per ICU")

# Show the plot
plt.show()
```

```python
# Create the data
num_hospitals = covid["hospitals"]
num_icu_beds = covid["icu_beds"]
names = covid["MMSA"]

# Define unique colors for each point based on the index
colors = [plt.cm.tab20(i) for i in range(len(names))]

#Increase plot size
fig, ax = plt.subplots(figsize=(15, 5))

ax.spines['top'].set_visible(False)


# Create the plot
scatter = plt.scatter(num_hospitals, num_icu_beds, c=range(len(names)), cmap="tab20")

# Set the plot title and axis labels
plt.title("Number of Hospitals vs. Number of ICU Beds")
plt.xlabel("Number of Hospitals")
plt.ylabel("Number of ICU Beds")

# Create legend
handles, labels = scatter.legend_elements(prop="colors")
legend = plt.legend(handles, names, loc="upper left", title="Name", fontsize="small")

# Set the colors of the legend labels to match the scatter plot
for i, label in enumerate(legend.get_texts()):
    label.set_color(colors[i])

# Show the plot
plt.show()
```
