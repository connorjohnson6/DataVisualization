```python
df = pd.read_csv('/Users/connorjohnson/Desktop/Data_300L/Week1/codon_usage.csv')
```

```python
#bar chart of kingdom counts
kingdoms = df['Kingdom'].unique()  #get the unique types of kingdoms from the kingdoms column
counts = df['Kingdom'].value_counts() #get the counts of each kingdom
dfplot = pd.DataFrame({'Kingdom': kingdoms, 'Count': counts})  #make a bar chart of each unique kingdom by count
ax = dfplot.plot.bar(x='Kingdom', y='Count', rot=0, title='Number of Species by Kingdom')
plt.show()
```

```python
#make a histogram of the data

just_codons = df.drop(df.columns[[1, 2, 3, 4, ]], axis=1)
just_codons.hist(bins=50)  #plots all the individual codons frequencies on their own axes
axes = just_codons.hist(bins=30,column='UUU',by='Kingdom') #plots just UUU frequencies by kingdom
plt.show()
```

```python
#Scatter Plot

plt.scatter(df.CAA, df.GAA, s=10)
plt.title("Scatter plot of Frequency of Glutamine Codons to Glutamic Acid Codons")
plt.xlabel('Glutamine - CAA')
plt.ylabel('Glutamic Acid - GAA')
plt.show()
```

```python
#Line plot

df.groupby('Kingdom')['Ncodons'].plot(legend=True)
plt.title("Line plot of Number of Codons by Kingdom")
plt.show()
```
